범주형 변수란?

범주형 변수를 인코딩 해야하는 이유
- 범주형 변수를 분류 모델에 넣으면 모델은 데이터를 인식하지 못고 에러 발생
- 이를 해결 하기 위해 범주형 변수를 수치형 변수로 인코딩 처리를 해줘야함
- 인코딩 방법의 대표적인 방식 :  Label Encoding / One - hot Encoding

```python
# school sex paid activities -> 범주형 변수
# pandas에서 get_dummies 함수 지원
# pd.get_dummies()는 one-hot encoding을 실행하는 함수임

df = pd.get_dummies(data = df, columns=['school', 'sex', 'paid', 'activities'],drop_first=True)
df.info()
```
